```python
#open all the files, try and parse the proper data, compare total uses of words or phrases

import os

#where all of our files were exported
lookingForClosed = False
startingPos = -1
endingPos = -1
totalFound = -1

condensedPage = ""

#for loop over all the files - for now we will just open one of the files
for i in range(0, 9999):
    try:
        f = open("indeed_files/" + str(i) + ".html", "r+")
    except FileNotFoundError:
        # doesn't exist
        break
    else:
        # exists
        #now we got a buncha data, time to figure out what is in it
        s = f.read()
        condensedPage= s
        #for q in range(len(s)):
        if (len(s) > 10):
            for q in range(len(s)):
                if (q < len(s)):
                    #this is where the magic happens!
                    #first, we will remove everything between "< >" to clear bad data
                    if (lookingForClosed):
                        if (s[q] == ">"):
                            totalFound = totalFound + 1
                            endingPos = q
                                #here's the issue - I am resetting s but keeping the "q" the same, even though the 
                                #full string is getting shorter every time. Not totally sure what to do 
                            condensedPage = condensedPage.replace(s[startingPos:endingPos+1],'')
                            #q = q - len(s[startingPos:endingPos+1])
                            #q = 0
                            lookingForClosed = False
                        else:
                            continue
                    elif (s[q] == "<"):
                        startingPos = q
                        lookingForClosed = True
                    #print(s[q])
                else:
                    continue
        #print (condensedPage)
        f = open("indeed_files/edited_files/" + str(i) + ".txt", "w+")
        f.write(condensedPage)
        f.close()
        print("edited " + str(i) + " to condensed .txt file")
```
